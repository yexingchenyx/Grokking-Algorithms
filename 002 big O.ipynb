```python
import numpy as np
import matplotlib.pyplot as plt

n = list(range(1, 100))
```

```python
plt.subplot()
plt.plot(n, [np.log2(v) for v in n])
plt.title('1: log2(n)')
plt.xlabel('n')
plt.ylabel('consume')
plt.show()

plt.subplot()
plt.plot(n, n)
plt.title('2: n')
plt.xlabel('n')
plt.ylabel('consume')
plt.show()

plt.subplot()
plt.plot(n, [v * np.log2(v) for v in n])
plt.title('3: nlog2(n)')
plt.xlabel('n')
plt.ylabel('consume')
plt.show()

plt.subplot()
plt.plot(n, [v ** 2 for v in n])
plt.title('4: n**2')
plt.xlabel('n')
plt.ylabel('consume')
plt.show()

plt.subplot()
plt.plot(n, [np.math.factorial(v) for v in n])
plt.title('5: n!')
plt.xlabel('n')
plt.ylabel('consume')
plt.show()
```
